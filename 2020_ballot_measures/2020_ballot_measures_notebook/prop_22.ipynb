## Campaign finances for Prop. 22, a CA 2020 ballot measure that would exempt rideshare drivers exempt from employee benefots

By Luke Harold

** current as of Sept. 26, 2020

```python
import pandas as pd
```

```python
committee_1 = pd.read_csv("prop_22_1.csv")
```

```python
committee_2 = pd.read_csv("prop_22_2.csv")
```

```python
committee_3 = pd.read_csv("prop_22_3.csv")
```

```python
committee_4 = pd.read_csv("prop_22_4.csv")
```

```python
committee_5 = pd.read_csv("prop_22_5.csv")
```

```python
contribs = pd.concat([committee_1, committee_2, committee_3, committee_4, committee_5], sort=True)
```

```python
contribs.groupby("position").amount.sum()
```

```python
top_supporters = contribs[contribs.position == "support"].sort_values("amount", ascending = False)
```

```python
top_supporters.groupby("name_of_contributor").amount.sum().reset_index()
```

```python
top_supporters.groupby("committee_name").amount.sum().reset_index()
```

```python
top_opponents = contribs[contribs.position == "oppose"].sort_values("amount", ascending = False)
```

```python
top_opponents.groupby("name_of_contributor").amount.sum().reset_index().sort_values("amount", ascending = False)
```

```python
top_opponents.groupby("committee_name").amount.sum().reset_index()
```
